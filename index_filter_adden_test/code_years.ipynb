```python
from openpyxl import load_workbook, Workbook

# Pfad zur Excel-Datei
datei_pfad = r"C:\Users\funke\OneDrive\Dokumente\Digital Humanities\Praxisprojekt\Alle_Daten_zusammen_tabelle.xlsx"
output_pfad = r"C:\Users\funke\OneDrive\Dokumente\Digital Humanities\Praxisprojekt\einzigartige_eintraege.xlsx"

# Excel-Datei laden
workbook = load_workbook(datei_pfad)
sheet = workbook.active  # Aktives Arbeitsblatt auswählen

# Spaltenwerte auslesen (z. B. Spalte A)
spalten_index = 5  # Index der Spalte (1-basiert, z. B. 1 für A, 2 für B)
werte = [row[0] for row in sheet.iter_rows(min_col=spalten_index, max_col=spalten_index, values_only=True)]

# Einzigartige Werte extrahieren
einzigartige_werte = list(set(werte))

# Neue Excel-Datei erstellen und speichern
neues_workbook = Workbook()
neues_sheet = neues_workbook.active

# Einzigartige Werte in die neue Datei schreiben
for i, wert in enumerate(einzigartige_werte, start=1):
    neues_sheet.cell(row=i, column=1, value=wert)

neues_workbook.save(output_pfad)

print(f"Die einzigartigen Werte wurden erfolgreich in '{output_pfad}' gespeichert.")

```

```python
# Liste der einzigartigen Werte
print("Liste der einzigartigen Werte:")
print(einzigartige_werte)

# Zugriff auf die Liste
for wert in einzigartige_werte:
    print(wert)
```
